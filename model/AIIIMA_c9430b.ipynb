```python
"""
dual_branch_psd_framework_final_with_gradcam.py

A corrected implementation using the user-preferred robust data loader,
adapted to generate both thermogram and PSD images for the dual-input model,
with Grad-CAM visualizations from the final fused features.

Framework Stages Implemented:
- Stage 1: PSD-Based Image Transformation (in data loader)
- Stage 2: Backbone-Based Feature Extraction (MobileNetV3 branch)
- Stage 3: Dual Parallel CNN Branches (for IR and PSD images)
- Stage 4: Simplified Gated Fusion (custom Keras layer)  <-- Grad-CAM target layer
- Stage 5: Final Classification (concatenation and MLP head)
- Grad-CAM: Visualize 10 test images using fused feature maps
"""
import os
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import (
    Layer, Input, Conv2D, MaxPooling2D, BatchNormalization, Activation,
    GlobalAveragePooling2D, Dense, Dropout, Concatenate
)
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.applications import MobileNetV3Small
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_curve, auc, roc_auc_score
import matplotlib.pyplot as plt

# ----------------------------------------------------
# Stage 1: PSD-Based Image Transformation
# ----------------------------------------------------
def create_psd(image):
    """Computes the log Power Spectral Density of a single-channel image."""
    f_transform = np.fft.fft2(image)
    f_transform_shifted = np.fft.fftshift(f_transform)
    power_spectrum = np.abs(f_transform_shifted) ** 2
    log_power_spectrum = np.log1p(power_spectrum)

    min_val, max_val = np.min(log_power_spectrum), np.max(log_power_spectrum)
    if max_val > min_val:
        log_power_spectrum = (log_power_spectrum - min_val) / (max_val - min_val)

    return log_power_spectrum.astype('float32')

def preprocess_and_load(img_path, size=(224, 224)):
    """Reads a thermogram, preprocesses it, and generates its PSD transformation."""
    try:
        img_gray = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        if img_gray is None:
            return None, None

        img_gray = cv2.medianBlur(img_gray, 3)
        img_resized = cv2.resize(img_gray, size)
        img_normalized = img_resized.astype('float32') / 255.0

        psd_img_gray = create_psd(img_normalized)

        thermogram_rgb = np.stack([img_normalized] * 3, axis=-1)
        psd_rgb = np.stack([psd_img_gray] * 3, axis=-1)

        return thermogram_rgb, psd_rgb
    except Exception as e:
        print(f"Error processing {img_path}: {e}")
        return None, None

# ----------------------------------------------------
# Corrected Data Loading Function
# ----------------------------------------------------
def load_dataset(data_dir=None, test_size=0.2, random_state=42):
    """
    Loads images using the robust directory search, adapted to create both
    thermogram and PSD pairs for the dual-input model.
    """
    if data_dir is None:
        data_dir = "/kaggle/input/thermal-images-for-breast-cancer-diagnosis-dmrir/Imagens e Matrizes da Tese de Thiago Alves Elias da Silva/Desenvolvimento da Metodologia"

    if not os.path.exists(data_dir):
        raise FileNotFoundError(f"data_dir not found: {data_dir}")

    # Robustly find category folders
    dir_contents = os.listdir(data_dir)
    doentes_folder = [f for f in dir_contents if "DOENTES" in f.upper() or "DOENTE" in f.upper()]
    saudaveis_folder = [f for f in dir_contents if "SAUD" in f.upper()]

    if not doentes_folder or not saudaveis_folder:
        raise ValueError("Could not find DOENTES or SAUDÁVEIS folders in the provided directory.")

    doentes_dir = os.path.join(data_dir, doentes_folder[0])
    saudaveis_dir = os.path.join(data_dir, saudaveis_folder[0])

    def load_images_from_path(directory, label):
        """Load all images from a given path and assign a label."""
        thermogram_list, psd_list, label_list = [], [], []
        for root, _, files in os.walk(directory):
            if "Segmentadas" in root:
                for file_name in files:
                    if file_name.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
                        img_path = os.path.join(root, file_name)
                        thermogram, psd = preprocess_and_load(img_path)
                        if thermogram is not None and psd is not None:
                            thermogram_list.append(thermogram)
                            psd_list.append(psd)
                            label_list.append(label)
        return thermogram_list, psd_list, label_list

    print("Loading cancer (DOENTES) images...")
    cancer_thermo, cancer_psd, cancer_labels = load_images_from_path(doentes_dir, 1)

    print("Loading healthy (SAUDÁVEIS) images...")
    healthy_thermo, healthy_psd, healthy_labels = load_images_from_path(saudaveis_dir, 0)

    # Combine lists from both classes
    X_thermogram = np.array(cancer_thermo + healthy_thermo, dtype=np.float32)
    X_psd = np.array(cancer_psd + healthy_psd, dtype=np.float32)
    y = np.array(cancer_labels + healthy_labels, dtype=np.int32)

    print(f"\nTotal samples loaded: {len(y)}")
    print(f"Thermogram data shape: {X_thermogram.shape}")
    print(f"PSD data shape: {X_psd.shape}")

    # Split dataset while keeping pairs aligned
    indices = np.arange(len(y))
    train_indices, test_indices = train_test_split(
        indices, test_size=test_size, random_state=random_state, stratify=y
    )

    X_train_ir, X_test_ir = X_thermogram[train_indices], X_thermogram[test_indices]
    X_train_psd, X_test_psd = X_psd[train_indices], X_psd[test_indices]
    y_train, y_test = y[train_indices], y[test_indices]

    print(f"Train shapes: IR={X_train_ir.shape}, PSD={X_train_psd.shape}")
    print(f"Test shapes: IR={X_test_ir.shape}, PSD={X_test_psd.shape}")

    return (X_train_ir, X_train_psd), (X_test_ir, X_test_psd), y_train, y_test

# ----------------------------------------------------
# Stage 4: Simplified Gated Fusion Layer
# ----------------------------------------------------
class GatedFusion(Layer):
    def __init__(self, channels, **kwargs):
        super(GatedFusion, self).__init__(**kwargs)
        self.channels = channels
        self.spatial_conv = Conv2D(channels, 1, activation='sigmoid')
        self.freq_conv = Conv2D(channels, 1, activation='sigmoid')

    def call(self, inputs):
        spatial, freq = inputs
        # Utilities
        u_spatial = self.spatial_conv(spatial)
        u_freq = self.freq_conv(freq)
        # Shapley-like normalized contributions
        w_spatial = u_spatial / (u_spatial + u_freq + 1e-8)
        w_freq = u_freq / (u_spatial + u_freq + 1e-8)
        # Weighted fusion
        fused = w_spatial * spatial + w_freq * freq
        return fused

    def get_config(self):
        config = super().get_config()
        config.update({"channels": self.channels})
        return config

# ----------------------------------------------------
# The Full Model Architecture
# ----------------------------------------------------
def create_cnn_branch(inputs, name_prefix):
    x = Conv2D(32, 3, padding='same')(inputs)
    x = BatchNormalization()(x); x = Activation('relu')(x); x = MaxPooling2D()(x)
    x = Conv2D(64, 3, padding='same')(x)
    x = BatchNormalization()(x); x = Activation('relu')(x); x = MaxPooling2D()(x)
    x = Conv2D(128, 3, padding='same')(x)
    x = BatchNormalization()(x); x = Activation('relu')(x); x = MaxPooling2D(name=f'{name_prefix}_out')(x)
    return x

def build_dual_branch_model(input_shape=(224, 224, 3)):
    """Builds the complete model based on the proposed framework."""
    ir_input = Input(shape=input_shape, name='ir_input')
    psd_input = Input(shape=input_shape, name='psd_input')

    # Backbone on IR
    backbone = MobileNetV3Small(input_tensor=ir_input, weights='imagenet', include_top=False, pooling='avg')
    backbone.trainable = False
    backbone_features = Dense(64, activation='relu', name='backbone_dense')(backbone.output)
    backbone_features = Dropout(0.4, name='backbone_dropout')(backbone_features)

    # Dual branches
    spatial_branch_out = create_cnn_branch(ir_input, 'spatial')
    frequency_branch_out = create_cnn_branch(psd_input, 'frequency')

    # Fused conv features (TARGET for Grad-CAM)
    fused_features = GatedFusion(channels=128, name="gated_fusion")([spatial_branch_out, frequency_branch_out])

    # Pool fused convs + concat with backbone vector
    x = GlobalAveragePooling2D(name='fusion_gap')(fused_features)
    final_features = Concatenate(name='final_concat')([x, backbone_features])

    # Head
    x = Dense(128, activation='relu', name='final_dense_1')(final_features)
    x = Dropout(0.5, name='final_dropout')(x)
    output = Dense(1, activation='sigmoid', name='classifier')(x)

    model = Model(inputs=[ir_input, psd_input], outputs=output, name='PSD_DualBranch_GatedFusion_Model')
    model.compile(optimizer=Adam(learning_rate=1e-3),
                  loss='binary_crossentropy',
                  metrics=['accuracy', tf.keras.metrics.AUC(name='auc')])
    return model

# ----------------------------------------------------
# Grad-CAM Utilities for Fused Features
# ----------------------------------------------------
def get_gradcam_heatmap_fused(model, img_ir, img_psd, layer_name="gated_fusion"):
    """
    Computes Grad-CAM heatmap using the final fused conv features.
    For binary sigmoid output, we use the score for class 1 directly.
    """
    grad_model = tf.keras.models.Model(
        [model.inputs],
        [model.get_layer(layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(
            [tf.expand_dims(img_ir, 0), tf.expand_dims(img_psd, 0)]
        )
        # predictions shape: (1, 1) for sigmoid binary classifier
        class_channel = predictions[:, 0]  # probability/logit for class 1

    # Gradients w.r.t. conv outputs
    grads = tape.gradient(class_channel, conv_outputs)  # shape: (1, H, W, C)
    conv_outputs = conv_outputs[0]                      # (H, W, C)
    grads = grads[0]                                    # (H, W, C)

    # Channel-wise importance: global average over spatial dims
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1))   # (C,)

    # Weighted sum of channels
    heatmap = tf.reduce_sum(conv_outputs * pooled_grads, axis=-1)  # (H, W)

    # ReLU and normalize to [0,1]
    heatmap = tf.maximum(heatmap, 0)
    denom = tf.reduce_max(heatmap) + 1e-8
    heatmap = heatmap / denom
    return heatmap.numpy()

def overlay_heatmap_on_image(heatmap, image_rgb, alpha=0.4):
    """
    Overlays a heatmap on the input image (expects image in [0,1], RGB).
    Returns BGR uint8 image suitable for cv2.imwrite or plt.imshow after RGB conversion.
    """
    # Resize heatmap to image size
    h, w = image_rgb.shape[:2]
    heatmap_resized = cv2.resize((heatmap * 255).astype(np.uint8), (w, h))

    # Apply color map
    heatmap_color = cv2.applyColorMap(heatmap_resized, cv2.COLORMAP_JET)  # BGR

    # Prepare base image (to BGR uint8)
    base_bgr = (image_rgb * 255.0).astype(np.uint8)
    base_bgr = cv2.cvtColor(base_bgr, cv2.COLOR_RGB2BGR)

    # Blend
    blended = cv2.addWeighted(base_bgr, 1 - alpha, heatmap_color, alpha, 0)
    return blended  # BGR

# ----------------------------------------------------
# Main Execution Block
# ----------------------------------------------------
if __name__ == '__main__':
    DATASET_PATH = "/kaggle/input/thermal-images-for-breast-cancer-diagnosis-dmrir/Imagens e Matrizes da Tese de Thiago Alves Elias da Silva/Desenvolvimento da Metodologia"

    if not os.path.exists(DATASET_PATH):
        print(f"Dataset path not found: {DATASET_PATH}")
    else:
        # Load
        (X_train_ir, X_train_psd), (X_test_ir, X_test_psd), y_train, y_test = load_dataset(data_dir=DATASET_PATH)

        # Build & train
        model = build_dual_branch_model()
        model.summary()

        callbacks = [
            EarlyStopping(monitor='val_loss', patience=100, restore_best_weights=True, verbose=1),
            ModelCheckpoint('best_gated_fusion_model.keras', monitor='val_accuracy', mode='max', save_best_only=True),
            ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6)
        ]

        history = model.fit(
            [X_train_ir, X_train_psd], y_train,
            validation_data=([X_test_ir, X_test_psd], y_test),
            epochs=100,
            batch_size=16,
            callbacks=callbacks,
            verbose=1
        )

        # Load best weights
        print("\nLoading best model for final evaluation...")
        model.load_weights('best_gated_fusion_model.keras')

        # --- Final Evaluation ---
        loss, acc, auc_val = model.evaluate([X_test_ir, X_test_psd], y_test, verbose=0)
        print(f"\n--- Final Model Evaluation ---")
        print(f"Test Loss: {loss:.4f}, Test Accuracy: {acc:.4f}, Test AUC: {auc_val:.4f}")

        # Predictions
        y_pred_proba = model.predict([X_test_ir, X_test_psd], batch_size=32).flatten()
        y_pred = (y_pred_proba > 0.5).astype(int)

        # Metrics
        acc = accuracy_score(y_test, y_pred)
        prec = precision_score(y_test, y_pred, zero_division=0)
        rec = recall_score(y_test, y_pred, zero_division=0)
        f1 = f1_score(y_test, y_pred, zero_division=0)
        print("\n===== Classification Metrics =====")
        print(f"Accuracy : {acc:.4f}")
        print(f"Precision: {prec:.4f}")
        print(f"Recall   : {rec:.4f}")
        print(f"F1-Score : {f1:.4f}")

        print("\nDetailed Classification Report:")
        print(classification_report(
            y_test, y_pred,
            target_names=['Healthy (0)', 'Cancer (1)'],
            zero_division=0
        ))

        # Confusion Matrix (matplotlib only, no seaborn dependency)
        cm = confusion_matrix(y_test, y_pred)
        print("\nConfusion Matrix:\n", cm)
        plt.figure(figsize=(6, 5))
        plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
        plt.title("Confusion Matrix")
        plt.colorbar()
        tick_marks = np.arange(2)
        plt.xticks(tick_marks, ['Healthy (0)', 'Cancer (1)'], rotation=45)
        plt.yticks(tick_marks, ['Healthy (0)', 'Cancer (1)'])
        thresh = cm.max() / 2.
        for i in range(cm.shape[0]):
            for j in range(cm.shape[1]):
                plt.text(j, i, format(cm[i, j], 'd'),
                         horizontalalignment="center",
                         color="white" if cm[i, j] > thresh else "black")
        plt.ylabel('True label')
        plt.xlabel('Predicted label')
        plt.tight_layout()
        plt.show()

        # ROC Curve
        fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
        roc_auc = auc(fpr, tpr)
        plt.figure(figsize=(6, 5))
        plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.4f})")
        plt.plot([0, 1], [0, 1], linestyle="--", color="gray")
        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")
        plt.title("Receiver Operating Characteristic (ROC)")
        plt.legend(loc="lower right")
        plt.tight_layout()
        plt.show()

        # ----------------------------------------------------
        # Grad-CAM on Final Fused Features (10 Images)
        # ----------------------------------------------------
        print("\nGenerating Grad-CAM visualizations from 'gated_fusion' on 10 test images...")
        os.makedirs("gradcam_results", exist_ok=True)

        import numpy as np
        import matplotlib.pyplot as plt
        import tensorflow as tf
        import cv2

        # ---- Grad-CAM function (using final fused features) ----
        def get_gradcam_heatmap(model, img_ir, img_psd, layer_name="gated_fusion", pred_index=None):
            grad_model = tf.keras.models.Model(
                [model.inputs],
                [model.get_layer(layer_name).output, model.output]
            )

            with tf.GradientTape() as tape:
                conv_outputs, predictions = grad_model([tf.expand_dims(img_ir, 0), tf.expand_dims(img_psd, 0)])
                if pred_index is None:
                    pred_index = tf.argmax(predictions[0])
                class_channel = predictions[:, pred_index]

            grads = tape.gradient(class_channel, conv_outputs)
            pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

            conv_outputs = conv_outputs[0]
            heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
            heatmap = tf.squeeze(heatmap)

            # Normalize heatmap to [0,1]
            heatmap = tf.maximum(heatmap, 0) / (tf.reduce_max(heatmap) + 1e-8)
            return heatmap.numpy()

        # ---- Overlay function ----
        def overlay_heatmap(heatmap, original_img, alpha=0.4):
            heatmap = cv2.resize(heatmap, (original_img.shape[1], original_img.shape[0]))
            heatmap = np.uint8(255 * heatmap)
            heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

            if original_img.shape[-1] == 1:
                original_img = np.squeeze(original_img)
                original_img = cv2.cvtColor(original_img, cv2.COLOR_GRAY2RGB)
            else:
                original_img = np.uint8(255 * original_img)

            superimposed_img = cv2.addWeighted(original_img, 1 - alpha, heatmap, alpha, 0)
            return superimposed_img

        # ---- Plot Original + GradCAM side by side ----
        def plot_gradcam_examples(model, X_test_ir, X_test_psd, y_test, num_samples=10, layer_name="gated_fusion"):
            sample_indices = np.random.choice(len(X_test_ir), size=num_samples, replace=False)

            plt.figure(figsize=(12, 2 * num_samples))
            for i, idx in enumerate(sample_indices):
                ir_img = X_test_ir[idx]
                psd_img = X_test_psd[idx]
                true_label = y_test[idx]

                # GradCAM heatmap
                heatmap = get_gradcam_heatmap(model, ir_img, psd_img, layer_name=layer_name)
                gradcam_img = overlay_heatmap(heatmap, ir_img)

                # Predictions
                pred_prob = model.predict([ir_img[np.newaxis], psd_img[np.newaxis]], verbose=0)[0][0]
                pred_label = int(pred_prob > 0.5)

                # Plot side by side
                plt.subplot(num_samples, 2, 2*i + 1)
                plt.imshow(np.squeeze(ir_img), cmap="gray")
                plt.title(f"Original IR\nTrue: {true_label}")
                plt.axis("off")

                plt.subplot(num_samples, 2, 2*i + 2)
                plt.imshow(gradcam_img)
                plt.title(f"GradCAM\nPred: {pred_label}, Prob: {pred_prob:.2f}")
                plt.axis("off")

            plt.suptitle("Original vs Grad-CAM on Final Fused Features", fontsize=16)
            plt.tight_layout()
            plt.show()

        # ---- Run visualization ----
        plot_gradcam_examples(model, X_test_ir, X_test_psd, y_test, num_samples=10, layer_name="gated_fusion")

```
